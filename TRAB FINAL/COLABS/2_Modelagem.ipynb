### PCA
- ha grupos parcialmente separados, mas com alguma sobreposi√ß√£o
- significa que os limites de decis√£o n√£o s√£o lineares
- Modelos lineares simples (como Regress√£o Logistica) podem n√£o capturar bem todas as rela√ßoes

- ajuste nos hiperparametros
  - Para modelos baseados em distancia (KNN, SVM)
    - SVM: kernel n√£o linear (rbf )
    - KNN: n_neighbors pequeno (5-10), pq ha separa√ßoes locais

  - Para modelos de √°rvores (Decision Tree, RF, GB)
    - clusters mostram rela√ßoes nao lineares: arvores lidam bem com isso
    - pode usar profundidades medias (max_depth=4-8) sem underfitting

### LDA
- ha sobreposi√ß√£o entre as classes, semelhante ao que foi visto no PCA.
- Alguns agrupamentos aparecem mais densos, mas as fronteiras continuam difusas e n√£o lineares

- ajuste nos hiperparametros
  - GB: ajustar o modelo de forma gradual
    - learning_rate baixo (0.05‚Äì0.1) e n_estimators mais alto (200‚Äì300), evitar overfitting
  - MLP: nao √© necess√°rio um modelo profundo
  - estrutura simples, com uma camada intermediaria moderada (hidden_layer_sizes=(50,)) e fun√ß√£o de ativa√ß√£o relu

```python

modelos = {
    "Regress√£o Log√≠stica": LogisticRegression(max_iter=1000, random_state=42),

    "KNN": KNeighborsClassifier(n_neighbors=7, weights='distance'),

    "Decision Tree": DecisionTreeClassifier(max_depth=6, random_state=42),

    "Random Forest": RandomForestClassifier(
        n_estimators=200,
        max_depth=6,
        random_state=42
    ),

    "SVM": SVC(kernel="rbf", C=1.0, gamma='scale', random_state=42),

    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=250,
        learning_rate=0.08,
        max_depth=4,
        random_state=42
    ),

    "MLP": MLPClassifier(
        hidden_layer_sizes=(50,),
        activation='relu',
        solver='adam',
        alpha=0.001,
        learning_rate_init=0.001,
        max_iter=1000,
        random_state=42
    )
}
```

```python
subconjuntos = {
    "Todas as Features": X,
    "Mutual Info (Top 10)": datasets["X_mi_top10"],
    "Mutual Info (Top 5)": datasets["X_mi_top5"],
    "Random Forest (Top 10)": datasets["X_rf_top10"],
    "Random Forest (Top 5)": datasets["X_rf_top5"],
    "RFE (Top 10)": datasets["X_rfe_top10"],
    "RFE (Top 5)": datasets["X_rfe_top5"]
}
```

```python
def plot_confusion_matrix(y_true, y_pred, model_name, subset_name, classes, output_dir):

    cm = confusion_matrix(y_true, y_pred)
    df_cm = pd.DataFrame(cm, index=classes, columns=classes)

    plt.figure(figsize=(6, 5))
    sns.heatmap(df_cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{model_name} - {subset_name}')
    plt.xlabel('Predito')
    plt.ylabel('Verdadeiro')


    safe_model_name = model_name.replace(" ", "_").replace("/", "_")
    safe_subset_name = subset_name.replace(" ", "_").replace("/", "_")
    filename = f"{output_dir}/cm_{safe_model_name}_{safe_subset_name}.png"
    plt.savefig(filename, bbox_inches='tight')
    plt.close()
    print(f"Matriz de confus√£o salva em: {filename}")
```

### TREINAR E SALVAR AS MATRIZES DE CONFUSAO

```python
classes = [0, 1, 2]

resultados = []
output_dir = './matrizes_confusao'
os.makedirs(output_dir, exist_ok=True)

for nome_sub, X_sub in subconjuntos.items():
    print(f"\nüîπ Subconjunto: {nome_sub} ({X_sub.shape[1]} features)")

    X_train_sub, X_test_sub, y_train_sub, y_test_sub = train_test_split(
        X_sub, y, test_size=0.2, random_state=42, stratify=y
    )

    for nome_modelo, modelo in modelos.items():
        modelo.fit(X_train_sub, y_train_sub)
        y_pred = modelo.predict(X_test_sub)

        acc = accuracy_score(y_test_sub, y_pred)
        f1 = f1_score(y_test_sub, y_pred, average="macro")

        resultados.append({
            "Subconjunto": nome_sub,
            "Modelo": nome_modelo,
            "Acur√°cia": acc,
            "F1 Score (macro)": f1
        })

        print(f"{nome_modelo:<20} | Acur√°cia: {acc:.3f} | F1: {f1:.3f}")

        plot_confusion_matrix(y_test_sub, y_pred, nome_modelo, nome_sub, classes, output_dir)


df_resultados = pd.DataFrame(resultados)
```

## VISUALIZA√á√ïES

```python
df_resultados = pd.DataFrame(resultados)
df_resultados = df_resultados.sort_values(by=["Acur√°cia"], ascending=False)

print("\nRESULTADOS FINAIS:")
display(df_resultados)
```

```python
melhores = df_resultados.groupby("Subconjunto").apply(lambda x: x.nlargest(1, "Acur√°cia"))
display(melhores[["Subconjunto", "Modelo", "Acur√°cia", "F1 Score (macro)"]])
```

```python
mapeamento_nomes = {
    "Todas as Features": "Todas",
    "Mutual Info (Top 10)": "MI (10)",
    "Mutual Info (Top 5)": "MI (5)",
    "Random Forest (Top 10)": "RF (10)",
    "Random Forest (Top 5)": "RF (5)",
    "RFE (Top 10)": "RFE (10)",
    "RFE (Top 5)": "RFE (5)"
}


df_resultados["Subconjunto_Abreviado"] = df_resultados["Subconjunto"].map(mapeamento_nomes)

modelos_unicos = df_resultados["Modelo"].unique()


n_modelos = len(modelos_unicos)
n_cols = 4 if n_modelos > 4 else n_modelos
n_rows = (n_modelos + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 8), tight_layout=True)
axes = axes.flatten()

for i, modelo in enumerate(modelos_unicos):
    ax1 = axes[i]

    df_modelo = df_resultados[df_resultados["Modelo"] == modelo]

    sns.barplot(
        data=df_modelo,
        x="Subconjunto_Abreviado",
        y="Acur√°cia",
        ax=ax1,
        color='gray'
    )
    ax1.set_title(f"{modelo}", fontsize=12)
    ax1.set_xlabel("")


    if i % n_cols != 0:
        ax1.set_ylabel("")
        ax1.set_yticklabels([])

    ax1.set_ylim(0.0, 1.0)
    ax1.tick_params(axis='y', labelcolor='black')


    ax2 = ax1.twinx()
    sns.pointplot(
        data=df_modelo,
        x="Subconjunto_Abreviado",
        y="F1 Score (macro)",
        ax=ax2,
        color='lightskyblue',
        linestyles="--",
        marker="o"
    )


    if (i + 1) % n_cols != 0:
        ax2.set_ylabel("")
        ax2.set_yticklabels([])

    ax2.set_ylim(0.0, 1.0)
    ax2.tick_params(axis='y', labelcolor='black')

    ax1.tick_params(axis='x', rotation=30)


for j in range(i + 1, len(axes)):
    axes[j].axis('off')


legend_elements = [
    plt.Rectangle((0, 0), 1, 1, fc='gray', ec='gray', label='Acur√°cia'),
    Line2D([0], [0], color='lightskyblue', lw=2, linestyle='--', marker='o', label='F1 Score')
]

fig.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.05, 0.95), title="M√©tricas")

plt.show()
```

- **1. Desempenho Geral e Compara√ß√£o entre Modelos/Subconjuntos**
  - Conclus√£o Geral: A Sele√ß√£o de Features n√£o apenas manteve a alta performance, mas em v√°rios casos, a melhorou em compara√ß√£o com o uso de Todas as Features.
  - Melhor Desempenho Absoluto:
    - O melhor resultado de acur√°cia foi obtido pelo MLP (Rede Neural) com o subconjunto MI (Top 10), alcan√ßando aproximadamente 0.909 (90.9%).
    - Outros subconjuntos como MI (Top 5), RFE (Top 10) e RFE (Top 5) tamb√©m alcan√ßaram resultados pr√≥ximos a 90.4% (principalmente com Regress√£o Log√≠stica).
  - Justificativa para o Desempenho Superior dos Subconjuntos:
    - A redu√ß√£o de features (de 21 para 5 ou 10) demonstrou ser altamente eficaz. Ao remover atributos redundantes ou de baixo valor informativo, o modelo se concentra apenas nos preditores mais fortes.
    - Diminui√ß√£o de Ru√≠do/Dimens√£o: Subconjuntos menores reduzem o ru√≠do e a complexidade desnecess√°ria (curse of dimensionality), o que permite que os modelos encontrem o padr√£o de estresse de forma mais eficiente e com menor risco de overfitting.

- **2. An√°lise do Baixo Desempenho de SVM e KNN com Todas as Features**
  - Os gr√°ficos mostram que o SVM e o KNN tiveram uma queda de desempenho dr√°stica quando usaram Todas as Features
  - A justificativa para esse colapso no desempenho est√° ligada √† forma como esses modelos operam e o efeito da Normaliza√ß√£o e da Irrelev√¢ncia das Features:
    - KNN (Baseado em Dist√¢ncia): √â extremamente sens√≠vel √† dimensionalidade e √† relev√¢ncia das features.
      - Irrelev√¢ncia: A inclus√£o de 21 features, muitas das quais t√™m pouca correla√ß√£o com o alvo (como mental_health_history ou study_load, conforme o Colab 1), faz com que a "dist√¢ncia" entre os pontos no espa√ßo 21-dimensional seja dominada por features que n√£o importam. O KNN n√£o consegue encontrar os vizinhos corretos para a classifica√ß√£o, resultando em desempenho pr√≥ximo ao de um palpite aleat√≥rio
    - SVM (Margem M√°xima): Embora o SVM com kernel 'rbf' seja robusto √† n√£o-linearidade, ele tamb√©m sofre com features irrelevantes.
      - Sobreposi√ß√£o no Hiperespa√ßo: Quando h√° muitas features irrelevantes, elas causam uma sobreposi√ß√£o de dados mesmo no hiperespa√ßo. O modelo tem dificuldade em encontrar uma fronteira clara, e o ajuste dos hiperpar√¢metros (C e Gamma) se torna ineficaz no espa√ßo de alta dimens√£o dominado pelo ru√≠do.
    - Conclus√£o: A Sele√ß√£o de Features foi cr√≠tica para SVM e KNN, pois filtraram o ru√≠do, tornando a m√©trica de dist√¢ncia e o c√°lculo do kernel significativos.

- **3. An√°lise da Performance da Regress√£o Log√≠stica (RL)**
  - A Regress√£o Log√≠stica (RL) alcan√ßou um desempenho muito bom nos subconjuntos MI (Top 5/10) e RFE (Top 5/10), o que parece contradizer a premissa de que o dataset n√£o √© linearmente separ√°vel (An√°lise PCA/LDA).
  - Justificativa:
    - A Escala dos Dados (Normaliza√ß√£o): RL √© um modelo linear, mas o uso da Normaliza√ß√£o MinMaxScale transformou todas as features em valores entre 0 e 1. Essa escala consistente pode ter facilitado o trabalho do modelo, permitindo que a RL encontrasse uma fronteira linear eficaz no espa√ßo das features escaladas (que √© diferente do espa√ßo projetado do PCA/LDA).
    - RFE e Linearidade: O m√©todo RFE (Recursive Feature Elimination) usou a pr√≥pria Regress√£o Log√≠stica como modelo base. Ao ser treinado repetidamente, o RFE selecionou o subconjunto de features que √© mais linearmente separ√°vel para a RL, otimizando inerentemente seu desempenho
    - Fronteira Otimizada: Embora o problema possa ser n√£o-linear globalmente, a RL, focando apenas nos preditores mais fortes (como blood_pressure e self_esteem), encontrou uma fronteira linear localmente muito eficaz para a classifica√ß√£o multi-classe.

  - Em resumo, a combina√ß√£o de pr√©-processamento (Normaliza√ß√£o) com a Sele√ß√£o de Features otimizada (RFE/MI) superou a limita√ß√£o te√≥rica da n√£o-linearidade para o modelo RL, impulsionando-o ao Top 5 dos melhores resultados.
